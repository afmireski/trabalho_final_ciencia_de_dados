```python
from __future__ import annotations

import argparse
import warnings
from pathlib import Path
from collections import Counter

import pandas as pd
```

```python

def infer_column_type(series: pd.Series) -> str:
    values = series.dropna()
    if values.empty:
        return "vazia"

    text_values = values.astype(str).str.strip()
    text_values = text_values[text_values != ""]
    if text_values.empty:
        return "vazia"

    if pd.api.types.is_bool_dtype(series):
        return "booleano"

    if pd.api.types.is_numeric_dtype(series):
        return "numérico"

    parsed_numeric = pd.to_numeric(text_values, errors="coerce")
    if parsed_numeric.notna().all():
        return "numérico"

    with warnings.catch_warnings():
        warnings.simplefilter("ignore", UserWarning)
        parsed_datetime = pd.to_datetime(text_values, errors="coerce", dayfirst=True)
    if parsed_datetime.notna().all():
        return "data/hora"

    lowered = text_values.str.lower()
    if lowered.isin({"true", "false", "yes", "no", "sim", "nao", "não", "0", "1"}).all():
        return "booleano"

    return "texto"
```

```python
path = './data/datatran2025.csv'

df = pd.read_csv(path, sep=None, engine="python", encoding="latin1")

print(f"Número de registros: {len(df)}")
print(f"Quantidade de colunas: {len(df.columns)}")

type_counts = Counter(infer_column_type(df[column_name]) for column_name in df.columns)

print("Quantidade por tipo:")
for column_type, count in sorted(type_counts.items()):
    print(f"- {column_type}: {count}")
```
